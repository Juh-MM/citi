```python
import pandas as pd
```

```python
from google.colab import files
# fazer o upload do arquivo
uploaded = files.upload()

# pega o nome do primeiro arquivo enviado
file_name = list(uploaded.keys())[0]

# le e armazena o arquivo CSV
base_despadronizada = pd.read_csv(file_name)
base_despadronizada.head(10)
```

#1 Padronizar Nivel_Senioridade:


```python
# 1. calcular a moda dos niveis
moda = base_despadronizada['Nivel_Senioridade'].mode()[0]

# 2. substituindo as variações das respostas
base_despadronizada['Nivel_Senioridade'] = base_despadronizada['Nivel_Senioridade'].replace({
    'Jr': 'Júnior',
    'JR': 'Júnior',
    'pleno': 'Pleno',
    'P': 'Pleno',
    'senior': 'Sênior',
    'N/D': moda
})

# 3. verificar se deu certo fazendo a contagem dos valores
base_despadronizada['Nivel_Senioridade'].value_counts()
```

#2. Padronizar Avaliacao_Tecnica e Avaliacao_Comportamental:


```python
# junta as colunas em uma variavel
colunas = ['Avaliacao_Tecnica', 'Avaliacao_Comportamental']

# converte para numero
for col in colunas:
    base_despadronizada[col] = pd.to_numeric(base_despadronizada[col], errors='coerce')

# calcula medias e preenche valores nulos
media_tecnica = round(base_despadronizada['Avaliacao_Tecnica'].mean(), 1)
media_comportamental = round(base_despadronizada['Avaliacao_Comportamental'].mean(), 1)

base_despadronizada['Avaliacao_Tecnica'].fillna(media_tecnica, inplace=True)
base_despadronizada['Avaliacao_Comportamental'].fillna(media_comportamental, inplace=True)

# converte para string e troca ponto por virgula
for col in colunas:
    base_despadronizada[col] = base_despadronizada[col].round(1).astype(str).str.replace('.', ',', regex=False)

# verifica resultado
base_despadronizada.head(13)
```

#3. Tratar Engajamento_PIGs:

```python
# transforma em str e remove o %
base_despadronizada['Engajamento_PIGs'] = base_despadronizada['Engajamento_PIGs'].astype(str).replace('%', '', regex=True)

# tranforma de volta para number
base_despadronizada['Engajamento_PIGs'] = pd.to_numeric(base_despadronizada['Engajamento_PIGs'], errors='coerce')

# transforma para valor descimal
decimal = (base_despadronizada['Engajamento_PIGs'])/100

# calcula a media e arredonda
media = round(decimal.mean(),1)

# preenche os campos NaN
decimal = decimal.fillna(media)

# atualiza a variavel
base_despadronizada['Engajamento_PIGs'] = decimal

# verificar se deu certo
base_despadronizada['Engajamento_PIGs'].head(20)
```

#4. Calcular  Score_Desempenho:

```python
# transformar , em . novamente para realizar calculos
base_despadronizada['Avaliacao_Tecnica'] = base_despadronizada['Avaliacao_Tecnica'].str.replace(',', '.', regex=False)
base_despadronizada['Avaliacao_Comportamental'] = base_despadronizada['Avaliacao_Comportamental'].str.replace(',', '.', regex=False)

# passar valores para number
Avaliacao_Tecnica = pd.to_numeric(base_despadronizada['Avaliacao_Tecnica'], errors='coerce')
Avaliacao_Comportamental = pd.to_numeric(base_despadronizada['Avaliacao_Comportamental'], errors='coerce')

# substituir valores nan por 0
base_despadronizada['Avaliacao_Tecnica'] = base_despadronizada['Avaliacao_Tecnica'].fillna(0)
base_despadronizada['Avaliacao_Comportamental'] = base_despadronizada['Avaliacao_Comportamental'].fillna(0)

# calcular score
Score_Desempenho = (Avaliacao_Tecnica * 0.5) + (Avaliacao_Comportamental * 0.5)

# verificar se deu certo
Score_Desempenho.head(10)

# adicionar na tabela original
base_despadronizada['Score_Desempenho'] = Score_Desempenho
base_despadronizada.head(20)
```

#5.  Criar nova coluna de Status_Membro:

```python
# funçao para definir o status do membro
def definir_status(row):
    if row['Score_Desempenho'] >= 7.0 and row['Engajamento_PIGs'] >= 0.8:
        return 'Em Destaque'
    else:
        return 'Padrão'

# aplicar a funçao a cada linha
base_despadronizada['Status_Membro'] = base_despadronizada.apply(definir_status, axis=1)

# verificar o resultado
base_despadronizada.head(20)
```

# Entrega

```python
base_padronizada = base_despadronizada

# fazer dowload do arquivo
base_padronizada.to_csv('base_padronizada.csv', index=False)
files.download('base_padronizada.csv')

base_padronizada.head(30)
```
